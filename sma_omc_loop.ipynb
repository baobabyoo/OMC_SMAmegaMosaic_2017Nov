## Loading the coordinate table

```python
infile = 'sma_pointings_smallcircle.coordinates'
rah = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=0
                )

ram = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=1
                )

ras = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=2
                )

decd = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=3
                )

decm = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=4
                )

decs = np.loadtxt(
                 infile, 
                 comments='#',
                 skiprows=0,
                 usecols=5
                )
```

## Evaluating RA and DEC

```python
ra  = ( rah + ram / 60.0 + ras / 3600.0 ) * 15.0
dec = np.arange( len(ra) ) * 0.0

for pnt_id in range( len(decd) ):
    if ( decd[pnt_id] >= 0.0 ):
        dec[pnt_id] = decd[pnt_id] + decm[pnt_id]/60.0 + decs[pnt_id]/3600.0
    else:
        dec[pnt_id] = decd[pnt_id] - decm[pnt_id]/60.0 - decs[pnt_id]/3600.0
        
for pnt_id in range( len(decd) ):
    print( pnt_id + 1, ra[pnt_id], dec[pnt_id] )
```

## Evaluating pointing offsets

```python
raoffset_arcsec  =  np.arange( len(ra) ) * 0.0
decoffset_arcsec =  np.arange( len(ra) ) * 0.0

for pnt_id in range( len(decd) ):
    raoffset_arcsec[pnt_id]  = ( ra[pnt_id]  - ra[0] )  * 3600.0 \
                               / math.cos( dec[pnt_id] * (math.pi / 180.0) )
    decoffset_arcsec[pnt_id] = ( dec[pnt_id] - dec[0] ) * 3600.0
    

    
for pnt_id in range( len(decd) ):
    print( pnt_id + 1, raoffset_arcsec[pnt_id],  decoffset_arcsec[pnt_id])
```

## Constructing Looping command

```python
# Constructing a dictionary between scan name and scan coordinate offset
offsetdict = {}
offsetdict['cal0'] = 'cal0'
offsetdict['cal1'] = 'cal1'
offsetdict['targ1'] = 'targ1'

for pnt_id in range(2, len(dec)+1 ):
    scan_name = 'targ' + str(pnt_id)
    offsetdict[scan_name] = '"m' + offsetdict['targ1'] \
                          + '[' + str( round(raoffset_arcsec[pnt_id  -1],2) ) + ','  \
                          +       str( round(decoffset_arcsec[pnt_id -1],2) ) + ']"'
    # print( scan_name, offsetdict[scan_name] )

loop1 = "&ObsLoop("
for scan_id in range( len(oldstyle_loop1_list) ):
   # print ( oldstyle_loop1_list[scan_id] 
   print( oldstyle_loop1_list[scan_id], ' -> ', offsetdict[ oldstyle_loop1_list[scan_id] ] )
   loop1 = loop1 + offsetdict[ oldstyle_loop1_list[scan_id] ]
   if ( scan_id != (len(oldstyle_loop1_list) -1) ):
      loop1 = loop1 + ','
   else:
      loop1 = loop1 + ');'

        
print( '\n\n' )        
        
loop2 = "&ObsLoop("
for scan_id in range( len(oldstyle_loop2_list) ):
   print( oldstyle_loop2_list[scan_id], ' -> ', offsetdict[ oldstyle_loop2_list[scan_id] ] )
   loop2 = loop2 + offsetdict[ oldstyle_loop2_list[scan_id] ]
   if ( scan_id != (len(oldstyle_loop2_list) -1) ):
      loop2 = loop2 + ','
   else:
      loop2 = loop2 + ');'
```
